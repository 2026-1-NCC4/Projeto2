**Gráficos:**

O gráfico abaixo representa a evolução do faturamento ao longo do tempo, permitindo observar possíveis variações nas vendas da empresa. No eixo X estão representados os períodos de tempo (meses), obtidos a partir da data de cada pedido registrada na base de dados. Já no eixo Y está representado o valor total faturado em cada mês, calculado pela soma da coluna totalamount, que corresponde ao valor monetário dos pedidos realizados. Esse tipo de visualização permite identificar tendências de crescimento ou queda nas vendas ao longo do tempo.

```python
# Evolucao do faturamento ao longo do tempo
storeOrder['scheduledat'] = pd.to_datetime(storeOrder['scheduledat'], errors='coerce')

faturamento_mensal = (
    storeOrder
    .dropna(subset=['scheduledat'])
    .groupby(storeOrder['scheduledat'].dt.to_period('M'))['totalamount']
    .sum()
)

plt.figure(figsize=(10, 5))
faturamento_mensal.plot(marker='o')
plt.title('Evolução do Faturamento ao Longo do Tempo')
plt.xlabel('Mês')
plt.ylabel('Faturamento Total')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

logo em seguida temos um gráfico que apresenta as 10 lojas com maior volume de faturamento dentro da base de dados da canoli. No eixo X está representado o valor total faturado por cada loja, calculado pela soma do campo totalamount, que representa o valor monetário dos pedidos realizados. No eixo Y estão listados os identificadores das lojas (storeid). Os valores apresentados correspondem ao faturamento acumulado de cada unidade, permitindo identificar quais lojas apresentam melhor desempenho em termos de vendas.

```python
# fundindo store e storeordder
storeOrderLoja = storeOrder.merge(
    store,
    left_on='storeid',
    right_on='id',
    how='left'
)
```

```python
#Top 10 lojas por faturameto

faturamento_loja = (
    storeOrderLoja
    .groupby('name', as_index=False)['totalamount']
    .sum()
    .sort_values(by='totalamount', ascending=False)
    .head(10)
)

plt.figure(figsize=(10, 6))
sns.barplot(data=faturamento_loja, x='totalamount', y='name')
plt.title('Top 10 Lojas por Faturamento')
plt.xlabel('Faturamento Total')
plt.ylabel('Nome')
plt.tight_layout()
plt.show()
```

o gráfico a seguir apresenta a distribuição da quantidade de pedidos realizados em cada canal de venda disponível na base de dados. No eixo X estão representados os diferentes canais de venda registrados na variável saleschannel, que podem incluir, por exemplo, aplicativo, site ou outros meios de compra. No eixo Y está representada a quantidade total de pedidos realizados em cada canal. Esses valores correspondem ao número de registros de pedidos associados a cada canal de venda, permitindo compreender quais meios são mais utilizados pelos clientes para realizar compras.

```python
pedidos_canal = storeOrder['saleschannel'].value_counts()

plt.figure(figsize=(8, 5))
sns.barplot(x=pedidos_canal.index, y=pedidos_canal.values)
plt.title('Distribuição dos Pedidos por Canal de Venda')
plt.xlabel('Canal de Venda')
plt.ylabel('Quantidade de Pedidos')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

O proximo gráfico apresenta a distribuição dos clientes da base de dados de acordo com o gênero. Cada fatia do gráfico representa a proporção de clientes pertencentes a cada categoria da variável gender. Os valores utilizados correspondem à quantidade de clientes registrados em cada gênero, permitindo visualizar a participação relativa de cada grupo dentro da base de clientes da empresa.

```python
# Destribuicao de clientes por genero
genero_clientes = customer['gender'].value_counts().sort_values()

plt.figure(figsize=(8,5))

sns.barplot(
    x=genero_clientes.values,
    y=genero_clientes.index
)

plt.title('Distribuição de Clientes por Gênero')
plt.xlabel('Quantidade de Clientes')
plt.ylabel('Gênero')

plt.tight_layout()
plt.show()
```

na grafico acima observa-se que, além das categorias M (masculino) e F (feminino), existem outros valores registrados na variável. Esses valores não correspondem a categorias de gênero válidas e provavelmente estão relacionados a inconsistências nos dados ou a campos que não foram preenchidos corretamente pelos usuários. Isso acontece porque as informações são coletadas diretamente de plataformas de pedidos utilizadas pela empresa, como aplicativos de delivery (por exemplo, iFood ou AnotaAi), onde o preenchimento de algumas informações cadastrais não é obrigatório ou não segue um padrão rígido.

**Seleção dos Dados**:

```python
store_order_sel = storeOrder[
    ['id', 'storeid', 'customerid', 'scheduledat',
     'totalamount', 'subtotalamount',
     'saleschannel', 'ordertype', 'status']
].copy()
```

o codigo acima seleciona apenas as colunas relevantes da tabela original, criando um novo DataFrame independente para análise, reduzindo complexidade e melhorando a performance.

**Limpeza e Padronização dos Dados**:

```python
store_order_sel['totalamount'] = pd.to_numeric(store_order_sel['totalamount'], errors='coerce')
store_order_sel['subtotalamount'] = pd.to_numeric(store_order_sel['subtotalamount'], errors='coerce')
```

o codigo acima converte os valores para formato numérico, substituindo valores inválidos por nulo (Nan), garantindo consistência nos cálculos.

```python
store_order_sel['scheduledat'] = pd.to_datetime(store_order_sel['scheduledat'], errors='coerce')
```

o codigo acima transforma a coluna de data em formato datetime, permitindo análises temporais. Valores inválidos são convertidos em nulo.

```python
store_order_sel = store_order_sel[
    store_order_sel['status'].isin([1,15,16,33,34,35,37,38])
]
```

o codigo acima filtra apenas pedidos com status válidos, garantindo que apenas vendas relevantes sejam consideradas.

```python
store_order_sel = store_order_sel.drop_duplicates()
```

o codigo acima remove registros duplicados para evitar distorções nas análises.

**Derivação dos Dados**:

```python
store_order_sel['ano'] = store_order_sel['scheduledat'].dt.year
store_order_sel['mes'] = store_order_sel['scheduledat'].dt.month
store_order_sel['dia'] = store_order_sel['scheduledat'].dt.day
store_order_sel.head(5)
```

o codigo acima cria novas colunas com informações de ano, mês e dia a partir da data, facilitando análises temporais.

**Integração dos Dados**:

```python
store_order_loja = store_order_sel.merge(
    store,
    left_on='storeid',
    right_on='id',
    how='left'
)
# display(store_order_loja)
store_order_loja.head(5)
```

o codigo acima realiza a junção entre pedidos e lojas, trazendo informações adicionais da loja para cada pedido

**Formatação dos Dados**

```python
faturamento_mensal = (
    store_order_sel
    .dropna(subset=['scheduledat'])
    .groupby(store_order_sel['scheduledat'].dt.to_period('M'))['totalamount']
    .sum()
)
faturamento_mensal.head(5)
```

o codigo acima agrupa os pedidos por mês e calcula o faturamento total, permitindo analisar a evolução ao longo do tempo

```python
vendas_canal = (
    store_order_sel
    .groupby('saleschannel')
    .agg(
        total_vendas=('totalamount', 'sum'),
        quantidade=('id', 'count')
    )
    .sort_values(by='total_vendas', ascending=False)
)
vendas_canal.head(5)
```

o codigo acima agrupa os dados por canal de venda, calculando o faturamento total e a quantidade de pedidos por canal.

```python
tipo_pedido = store_order_sel['ordertype'].value_counts(normalize=True) * 100
display(tipo_pedido.round(2))
```

o codigo acima calcula a porcentagem de cada tipo de pedido, permitindo entender a distribuição entre as categorias.
